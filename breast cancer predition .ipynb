# APPLYING CLASSIFICATION BASED MACHINE LEARNING MODELS AND DEEP LEARNING MODELS AS WELL

## splitting the data into test and train

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=0)
```

## applying standard scalar to the data

```python
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
final_accuracy=[]
model_names=[]
```

## Gaussian naive bayes algorithm

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
print(model)
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,y_pred)
sns.heatmap(cm,cmap='RdYlGn',linewidths=0.30)
plt.show()
acnb=accuracy_score(y_test,y_pred)
print("accuracy :",acnb)
final_accuracy.append(acnb)
model_names.append("naive bayes")
print(classification_report(y_test,y_pred))
```

## Logistic regression

```python
from sklearn import linear_model
model=linear_model.LogisticRegression()
print(model)
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,y_pred)
sns.heatmap(cm,cmap='RdYlGn',linewidths=0.30)
plt.show()
aclr=accuracy_score(y_test,y_pred)
print("accuracy :",aclr)
final_accuracy.append(aclr)
model_names.append("logistic regression")
print(classification_report(y_test,y_pred))
```

## KNN

```python
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier(n_neighbors=1)
print(model)
ac=[]
for i in range(1,50):
  z=KNeighborsClassifier(n_neighbors=i)
  z.fit(x_train,y_train)
  y_pred=z.predict(x_test)
  aci=accuracy_score(y_test,y_pred)
  ac.append(aci)
maxac=np.where(ac==max(ac))
for i in range(0,len(maxac[0])):
  a=maxac[0][i]
  print(a)
  z=KNeighborsClassifier(n_neighbors=a)
  z.fit(x_train,y_train)
  y_pred=z.predict(x_test)
  from sklearn.metrics import confusion_matrix,accuracy_score
  cm=confusion_matrix(y_test,y_pred)
  sns.heatmap(cm,cmap='RdYlGn',linewidths=0.30)
  plt.show()
  acknn=accuracy_score(y_test,y_pred)
  print("accuracy :",acknn)
  print(classification_report(y_test,y_pred))
  final_accuracy.append(acknn)
  b="knn"+str(a)
  model_names.append(b)
```

## decision tree

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
print(model)
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,y_pred)
sns.heatmap(cm,cmap='RdYlGn',linewidths=0.30)
plt.show()
acdt=accuracy_score(y_test,y_pred)
print("accuracy :",aclr)
print(classification_report(y_test,y_pred))
final_accuracy.append(acdt)
model_names.append("decision tree")
tree.plot_tree(model,feature_names=x.columns)
```

## random forest

```python
from sklearn.ensemble import RandomForestClassifier  
model=RandomForestClassifier(n_estimators=10,criterion="entropy")  
print(model)
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,y_pred)
sns.heatmap(cm,cmap='RdYlGn',linewidths=0.30)
plt.show()
acrf=accuracy_score(y_test,y_pred)
print("accuracy :",acrf)
print(classification_report(y_test,y_pred))
final_accuracy.append(acrf)
model_names.append("random forest")
from sklearn.tree import plot_tree
# plt.figure(figsize=(80,40))
plot_tree(model.estimators_[9],feature_names=x.columns,class_names=["0","1"],filled=True);
```

## support vector classifier

```python
from sklearn.svm import SVC
model=SVC(kernel='linear',random_state=0)  
print(model)
model.fit(x_train, y_train)
y_pred=model.predict(x_test)
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,y_pred)
sns.heatmap(cm,cmap='RdYlGn',linewidths=0.30)
plt.show()
acsv=accuracy_score(y_test,y_pred)
print("accuracy :",acsv)
print(classification_report(y_test,y_pred))
final_accuracy.append(acsv)
model_names.append("support vector")
```

## XG Boost

```python
from xgboost import XGBClassifier
model=XGBClassifier()
print(model)
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,y_pred)
sns.heatmap(cm,cmap='RdYlGn',linewidths=0.30)
plt.show()
acxgb=accuracy_score(y_test,y_pred)
print("accuracy :",acxgb)
print(classification_report(y_test,y_pred))
final_accuracy.append(acxgb)
model_names.append("XGVoost")
```

## Deep learning model

```python
from numpy import loadtxt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
model = Sequential()
model.add(Dense(12,input_shape=(30,),activation='relu'))
model.add(Dense(8,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(x_train,y_train,epochs=100,batch_size=10)
_, accuracy=model.evaluate(x_test,y_test)
print('Accuracy: %.2f' % (accuracy*100))
print(classification_report(y_test,y_pred))
final_accuracy.append(accuracy)
model_names.append("deep learning 1")
```

```python
from numpy import loadtxt
from sklearn.svm import SVC
from matplotlib import pyplot
import joblib
import sys
sys.modules['sklearn.externals.joblib']=joblib
from mlxtend.feature_selection import SequentialFeatureSelector as sfs
```

## subset selection for support vector classifier

```python
model=SVC()
print(model)
sfs1=sfs(estimator=model,k_features='best',forward=True,verbose=1,scoring='r2',cv=3)
sfs1=sfs1.fit(x,y)
t=sfs1.subsets_
t=pd.DataFrame(t)
t=t.transpose()
t
scr=t['avg_score']
minn=scr.min()
maxx=scr.max()
sns.barplot(np.arange(t.shape[0]),scr)
plt.show()
ind_scr=t['cv_scores']
ind_scr=np.array(ind_scr)
ind_scr_1=[]
ind_scr_2=[]
ind_scr_3=[]
for i in range(0,len(ind_scr)):
    ind_scr_1.append(ind_scr[i][0])
    ind_scr_2.append(ind_scr[i][1])
    ind_scr_3.append(ind_scr[i][2])
ax = plt.subplot(111)
ax.bar(np.arange(t.shape[0])-0.2, ind_scr_1, width=0.2, color='b', align='center')
ax.bar(np.arange(t.shape[0])    , ind_scr_2, width=0.2, color='g', align='center')
ax.bar(np.arange(t.shape[0])+0.2, ind_scr_3, width=0.2, color='r', align='center')
plt.show()
scr=np.array(scr)
for i in range(0,len(scr)-1):
    if scr[i]>=scr[i+1]:
        break
a=t['feature_names']
a=np.array(a)
print('the model will work best with',i+1,'columns and the columns are : ',a[i])
print(t)
z=np.array(a[3])
dfz=pd.DataFrame(z)
dfz.to_csv('best_columns.csv')
dfz
```
